##Splitting Data into Train and Test sets

```
from sklearn.model_selection import train_test_split
```

```
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

##Label Encoder for Target
It is used to convert the labels of 'status' to numeric form.

```
from sklearn.preprocessing import LabelEncoder
```

```
le = LabelEncoder()
```

```
lyt = le.fit_transform(y_train)
```

```
lyt
```

```
y_train
```

```
lys = le.fit_transform(y_test)
```

```
lys
```

```
y_test
```

##Standardization
used to get better efficiency.

```
from sklearn.preprocessing import StandardScaler
```

```
sc = StandardScaler()
```

```
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

##Train on different algorithms

```
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```
l_cla = LogisticRegression()
k_cla = KNeighborsClassifier()
d_cla = DecisionTreeClassifier()
r_cla = RandomForestClassifier()
s_cla = SVC(kernel='linear')
ks_cla = SVC(kernel='rbf')
```

```
l_cla.fit(x_train, lyt)
k_cla.fit(x_train, lyt)
d_cla.fit(x_train, lyt)
r_cla.fit(x_train, lyt)
s_cla.fit(x_train, lyt)
ks_cla.fit(x_train, lyt)
```

```
l_pred = l_cla.predict(x_test)
k_pred = k_cla.predict(x_test)
d_pred = d_cla.predict(x_test)
r_pred = r_cla.predict(x_test)
s_pred = s_cla.predict(x_test)
ks_pred = ks_cla.predict(x_test)
```

```
from sklearn.metrics import confusion_matrix
```

```
l_c = confusion_matrix(lys, l_pred)
k_c = confusion_matrix(lys, k_pred)
d_c = confusion_matrix(lys, d_pred)
r_c = confusion_matrix(lys, r_pred)
s_c = confusion_matrix(lys, s_pred)
ks_c = confusion_matrix(lys, ks_pred)
```
